```python
import csv
f = open('guns.csv', 'r')
data = list(csv.reader(f))
data[:5]
```

```python
headers = data[0]
headers
```

```python
data = data[1:]
data[:5]
```

```python
years = [year[1] for year in data]
year_counts = {}
for year in years:
    if year in year_counts:
        year_counts[year] += 1
    else:
        year_counts[year] = 1
year_counts
```

```python
import datetime
dates = [datetime.datetime(int(item[1]),int(item[2]),day=1) for item in data]
dates[:5]
```

```python
date_counts = {}
for date in dates:
    if date in date_counts:
        date_counts[date] += 1
    else:
        date_counts[date] = 1
        
date_counts
```

```python
sex_counts = {}
for item in data:
    if item[5] in sex_counts:
        sex_counts[item[5]] += 1
    else:
        sex_counts[item[5]] = 1
sex_counts
```

```python
race_counts = {}
for item in data:
    if item[-4] in race_counts:
        race_counts[item[-4]] += 1
    else:
        race_counts[item[-4]] = 1
race_counts
```

```python
census = open('census.csv','r')
census = list(csv.reader(census))
census
```

```python
mapping = {}
mapping['Asian/Pacific Islander'] = census[1][-2] + census[1][-3]
mapping['Black'] = census[1][-5]
mapping['Native American/Native Alaskan'] = census[1][-4]
mapping['Hispanic'] = census[-1][-6]
mapping['White'] = census[-1][-7]
mapping
```

```python
race_per_hundredk = {}

for key in race_counts:
    race = (race_counts[key]/int(mapping[key])) * 100000
    race_per_hundredk[key] = race
    
race_per_hundredk
```

```python
races = [item[-4] for item in data]
races
```

```python
homicide_race_counts = {}
for index, intent in enumerate(intents):
    if intent == 'Homicide':
        if races[index] in homicide_race_counts:
            homicide_race_counts[races[index]] += 1
        else:
            homicide_race_counts[races[index]] = 1
            
            
homicide_race_counts
```

```python
homicide_per_hundredk = {}

for key in homicide_race_counts:
    hom = (homicide_race_counts[key]/int(mapping[key])) * 100000
    homicide_per_hundredk[key] = hom
    
homicide_per_hundredk
```
